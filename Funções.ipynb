### alternar P_ID entre longo e curto

```python
def converte(processo): # valida o tamanho
    ''' 
    esta função alterna o formato de P_ID entre longo e curto.
    bloco1-bloco2 == ordinal do processo e código verificador
    bloco3 == ano de início do processo
    bloco4 == órgão ou segmento do Judiciário (ex.: 1 para Justiça Estadual, 2 para Justiça Federal, 3 para Justiça do Trabalho etc.).
    bloco5 == Tribunal ou circunscrição judiciária responsável (ex.: 02 = TRT da 2ª Região, 26 = TJSP)
    bloco6 == Unidade de origem (vara, juizado ou seção judiciária onde o processo foi distribuído)
    '''
    if len(processo) != 20 and len(processo) != 25: 
        print('O código não é válido') 
        return None # encerra se não for válido 
        # se for longo (25 caracteres), remove separadores 
    if len(processo) == 25: 
        novo = processo.replace('-', '').replace('.', '') 
    else: # se for curto (20 caracteres), insere separadores 
        novo = (processo[:7] + "-" + processo[7:9] + "." + processo[9:13] + "." 
        + processo[13]+'.'+processo[14:16] + "." + processo[16:]) 
    return novo 
            
# exemplo 
processo = '0078100-88.2009.5.02.0029' 
print(converte(processo)) # retorna versão curta (sem separadores) 
processo_curto = '00781008820095020029' 
print(converte(processo_curto)) # retorna versão longa (com separadores)
```

### criar uma lista Python com o nome de todos os arquivos de um diretório

```python
import os
from tqdm.notebook import tqdm
import time

def lst_dir(diretorio): 
    ''' o diretório deve conter o caminho (path). 
    Todos os arquivos do diretório serão listados, as pastas não.
    '''
    lista=[]
    
    for p in tqdm(os.listdir(diretorio)):
        time.sleep(.05)
        pp=os.path.join(diretorio,p)
        if os.path.isfile(pp):
            lista.append(p)
    return lista

# exemplo
pasta='E:/Pyproj/Ceneviva/dir_P'
print(lst_dir(pasta))
print('\nnúmero de arquivos no diretório: ',len(lst_dir(pasta)))
```

### Fazer o backup 'destino' de um arquivo 'origem'

```python
# ambiente
import os
import shutil

def bkp(origem,destino): # incluir o caminho no nome dos arquivos
    try:
        if os.path.exists(origem):
            shutil.copyfile(origem, destino)
            print("Backup criado com sucesso!")
        else:
            print("Arquivo original não encontrado.")
    except Exception as e:
        print(f"Ocorreu um erro ao copiar o arquivo: {e}")
    return

# example
bkp('E:/Ceneviva/proj/teste.xlsx','E:/Ceneviva/proj/teste_bkp.xlsx')


```

### Ler o número de páginas de um arquivo pdf

```python
# ambiente
import PyPDF2

def Pg_pdf(caminho_arquivo):
    
    '''ler o número de páginas de um arquivo pdf'''
    
    with open(caminho_arquivo, "rb") as arquivo:
        leitor = PyPDF2.PdfReader(arquivo)
        Pag = len(leitor.pages)
        return Pag

# Exemplo de uso:
arquivo_pdf = 'E:/Ceneviva/Dir_P/1500025-65.2017.8.26.0511.pdf'
print(f"O PDF tem {Pg_pdf(arquivo_pdf)} páginas.")
```
